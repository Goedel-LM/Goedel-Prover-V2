```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set a Seaborn theme
sns.set_theme()

# --- Data ---
revision_iterations = np.array([32, 64, 128, 256, 512, 1024])
our_32B_revision_performance = np.array([90.36, 91.39, 91.90, 92.31, 92.41, 92.62])

iterations = np.array([32, 64, 128, 256, 512, 1024, 2048, 4096, 8192])
our_32B_performance =  np.array([88.11, 89.75, 90.50, 90.96, 91.55, 91.80, 92.00, 92.20, 92.20])
our_8B_performance = np.array([83.3, 84.0, 84.8, 85.2, 85.7, 86.5, 87.3, 87.7, 88.1])

dpsk_iterations = np.array([32, 1024, 8192])
dpsk_671B_performance = np.array([82.4, 86.6, 88.9])
dpsk_7B_performance = np.array([75.6, 79.9, 82.0])

kimina_iterations = np.array([32, 1024])
kimina_72B_performance = np.array([84.0, 87.7])


# ===================================================================
# THE SOLUTION: Apply the y-axis logit transformation
# ===================================================================

# 1. Define a helper function for the logit transformation
def logit_transform(p):
    """Transforms percentage p (0-100) to the logit scale log(p/(1-p))."""
    p_frac = np.asarray(p) / 100.0
    # Use np.clip to avoid errors at p=0 or p=100
    epsilon = 1e-7
    p_frac = np.clip(p_frac, epsilon, 1 - epsilon)
    return np.log(p_frac / (1 - p_frac))


# --- Plotting ---
plt.figure(figsize=(5.8, 5.8))
ax = plt.gca() # Get current axes

# 2. Plot the data with LOGIT-TRANSFORMED y-values
# The x-axis calculations for "Effective Compute" are preserved
plt.plot(7/8*dpsk_iterations, logit_transform(dpsk_7B_performance), label='Deepseek-Prover-V2-7B', marker='o', linestyle='-', color="firebrick")
plt.plot(67/8*dpsk_iterations, logit_transform(dpsk_671B_performance), label='Deepseek-Prover-V2-671B', marker='o', linestyle='-',  color="sandybrown")
plt.plot(72/8*kimina_iterations, logit_transform(kimina_72B_performance), label='Kimina-Prover-72B', marker='o', linestyle='-', color="saddlebrown")
plt.plot(8/8*iterations, logit_transform(our_8B_performance), label='Goedel-Prover-V2-8B', marker='o', linestyle='-', color="mediumseagreen")
plt.plot(32/8*iterations, logit_transform(our_32B_performance), label='Goedel-Prover-V2-32B', marker='o', linestyle='-', color="cornflowerblue")
plt.plot(32/8*revision_iterations, logit_transform(our_32B_revision_performance), label='Goedel-Prover-V2-32B (self-correction)', marker='o', linestyle='-', color="mediumslateblue")


# --- Axis Configuration ---
plt.xscale('log')
plt.xlim(150/8, 700000/8)

# 3. Set y-axis to a logit scale
# Define the desired tick labels in the original percentage scale
y_tick_labels_pct = np.array([70, 80, 85, 90, 92])
# Calculate the positions of these ticks on the new logit-transformed scale
y_tick_values_logit = logit_transform(y_tick_labels_pct)
# Set the y-axis ticks to the logit positions, but label them with the original percentages
plt.yticks(y_tick_values_logit, labels=[f'{val}' for val in y_tick_labels_pct], fontsize=18)

# 4. Set the y-axis limits on the logit scale (corresponds to original 65% to 95%)
plt.ylim(logit_transform(62), logit_transform(93))

# --- Titles, Labels, and Layout ---
plt.title('Performance vs Inference Compute', fontsize=18)
plt.ylabel('Performance on miniF2F (%)', fontsize=18)
plt.xlabel('Effective Compute = Passes × Model Size', fontsize=16, labelpad=15)

# --- Final Touches ---
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12, loc='lower right') # Moved legend for better visibility
plt.tight_layout()

# Save and show the plot
plt.savefig("inference_scale_performance_compute_adjusted.png", dpi=300)
plt.show()
```
